```python

def extract_data(extracted_data, data):
    for index in range(len(data)):
        el = data[6][index]
        json_load = (json.loads(el))
        if(len(json_load.keys()) == 1):
            for key in json_load.keys():
                inner_element = json_load[key]
                if('selectedText' in inner_element.keys()):
                    if(inner_element['startIndexInSegment'] != -1):
                        extracted_data.append([inner_element['selectedText'], data[5][index]])
        if(len(json_load.keys()) > 1):
            text_data = ''
            startIndex = 0
            endIndex = 0
            for key in json_load.keys():
                inner_element = json_load[key]
                if('selectedText' in inner_element.keys()):
                    if(inner_element['startIndexInSegment'] != -1):
                        if(startIndex == 0 and endIndex == 0):
                            startIndex = inner_element['startIndexInSegment']
                            endIndex = inner_element['endIndexInSegment']
                            text_data = inner_element['selectedText']
                            
                        elif(inner_element['startIndexInSegment'] < startIndex and
                          inner_element['endIndexInSegment'] < startIndex):
                            text_data = inner_element['selectedText'] + " " + text_data
                            startIndex = inner_element['startIndexInSegment']
                            
                        elif(inner_element['startIndexInSegment'] < startIndex and
                            inner_element['endIndexInSegment'] > startIndex and
                            inner_element['endIndexInSegment'] < endIndex):
                            pos = inner_element['startIndexInSegment'] - startIndex
                            text_data = inner_element['selectedText'][:pos] + " " + text_data
                            startIndex = inner_element['startIndexInSegment']
                            
                        elif(inner_element['startIndexInSegment'] < startIndex and
                            inner_element['startIndexInSegment'] < endIndex and
                            inner_element['endIndexInSegment'] > endIndex):
                            pos = endIndex - inner_element['startIndexInSegment'] + 1
                            text_data = text_data + " " + inner_element['selectedText'][pos:]
                            endIndex = inner_element['endIndexInSegment']
                            
                        elif(inner_element['startIndexInSegment'] > endIndex and
                          inner_element['endIndexInSegment'] > endIndex):
                            text_data = text_data + " " + inner_element['selectedText']
                            endIndex = inner_element['endIndexInSegment']
                            
                        elif(inner_element['startIndexInSegment'] < startIndex and
                          inner_element['endIndexInSegment'] > endIndex):
                            startIndex = inner_element['startIndexInSegment']
                            endIndex = inner_element['endIndexInSegment']
                            text_data = inner_element['selectedText']
                            
            if(text_data != ''):
                extracted_data.append([text_data, data[5][index]])
```

```python
import pandas as pd
import json
import glob

glued_data = pd.DataFrame()
directoryPath = 'annotations/'
extracted_data = []
for file_name in glob.glob(directoryPath + '*.csv'):
    data = pd.read_csv(file_name, low_memory=False, header=None)
    extract_data(extracted_data, data)
```

```python
header = ["text", "class"]
wrangled_data = pd.DataFrame(extracted_data, columns=header)
wrangled_data.drop_duplicates(subset ="text",
                     keep = 'first', inplace = True)
wrangled_data
```

```python
wrangled_data.to_csv('dataset/wrangled_data.csv', index=False)
```
