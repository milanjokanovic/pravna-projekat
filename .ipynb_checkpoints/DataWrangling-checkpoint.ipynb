```python
import pandas as pd

data = pd.read_csv("annotations/20_theatlantic.com.csv", header=None)
data[6][4]
```

```python
for index in range(len(data)):
    el = data[6][index]
    print(el.find('selectedText'))
    startIndex = el.find('selectedText') + 16
    #endIndex = el.find('')
    extractedElement = el[startIndex:]
    endIndex = extractedElement.find('value') - 4
    startValueIndex = extractedElement.find('value') + 9
    endValueIndex = len(extractedElement) - 5 
    print(extractedElement)
    print('\n')
    print(extractedElement[startValueIndex:endValueIndex])
    print('\n')
    print(data[5][index])
    print('\n')
```
