```python
def extract_data(extracted_data, data):
    for index in range(len(data)):
        el = data[6][index]
        json_load = (json.loads(el))
        if(len(json_load.keys()) == 1):
            for key in json_load.keys():
                inner_element = json_load[key]
                if('selectedText' in inner_element.keys()):
                    extracted_data.append([inner_element['selectedText'], data[5][index]])
        if(len(json_load.keys()) > 1):
            text_data = ''
            for key in json_load.keys():
                inner_element = json_load[key]
                if('selectedText' in inner_element.keys()):
                    text_data = text_data + " " + inner_element['selectedText']
            if(text_data != ''):
                extracted_data.append([text_data, data[5][index]])
```

```python
import pandas as pd
import json
import glob

glued_data = pd.DataFrame()
directoryPath = 'annotations/'

extracted_data = []
for file_name in glob.glob(directoryPath + '*.csv'):
    data = pd.read_csv(file_name, low_memory=False, header=None)
    extract_data(extracted_data, data)
#extracted_data = []
#data = pd.read_csv("annotations/20_theatlantic.com.csv", header=None)
```

```python
header = ["text", "class"]
wrangled_data = pd.DataFrame(extracted_data, columns=header)
wrangled_data.drop_duplicates(subset ="text",
                     keep = 'first', inplace = True)
wrangled_data
```
